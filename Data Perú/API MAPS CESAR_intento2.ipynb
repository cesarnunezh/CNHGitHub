### 2. Importing the data set of the Health Establishments in Cajamarca

```python
import pandas as pd
import requests
import time
import os
import matplotlib.pyplot as plt
import seaborn as sns
import math
from geopy.distance import geodesic
```

```python
def distancia_euclidiana(lat1, lon1, lat2, lon2):
    coord1 = (lat1, lon1)
    coord2 = (lat2, lon2)
    return geodesic(coord1, coord2).kilometers

# Ejemplo de uso
lat1, lon1 = 40.7128, -74.0060  # Coordenadas de Nueva York
lat2, lon2 = 34.0522, -118.2437  # Coordenadas de Los Ángeles
distancia = distancia_euclidiana(lat1, lon1, lat2, lon2)
print("La distancia euclidiana entre Nueva York y Los Ángeles es: {:.2f} kilómetros".format(distancia))
```

### 5. Using Google API Directions to obtain the distance and travel time to the closest health establishment for each district

```python
HE_Cajamarca_min["travel_time_best_guess"] = pd.Series(dtype=float)
HE_Cajamarca_min["travel_distance_best_guess"] = pd.Series(dtype=float)

models = ["best_guess"]

for index,row in HE_Cajamarca_min.iterrows():
    lat1 = HE_Cajamarca_min.loc[index,'LATITUD']
    lng1 = HE_Cajamarca_min.loc[index,'LONGITUD']
    origin = f'{round(lat1, 5)},{round(lng1, 5)}'
    print(origin)
    lat2 = HE_Cajamarca_min.loc[index,'LATITUD.1']
    lng2 = HE_Cajamarca_min.loc[index,'LONGITUD.1']
    destination = f'{round(lat2, 5)},{round(lng2, 5)}'
    print(destination)

    for model in models:
        try:
            req2 = requests.get( f"https://maps.googleapis.com/maps/api/directions/json?origin={origin}&destination={destination}&departure_time=now&traffic_model={model}&mode=driving&region=pe&key=AIzaSyBP4skpBaY8ttDvKzik68cpekzD9cV_1cM")
            data_json = req2.json()
            HE_Cajamarca_min.loc[index, f'travel_time_{model}'] = data_json['routes'][0]['legs'][0]['duration']['value']
            HE_Cajamarca_min.loc[index, f'travel_distance_{model}'] = data_json['routes'][0]['legs'][0]['distance']['value']
            time.sleep(0.1)
        except:
            continue
```
